```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np 
import matplotlib.pyplot as plt

#data set yaratalım.

X= np.sort(5 * np.random.rand(500,1), axis=0)
y= np.sin(X).ravel()

#5 adımda bir olacak şekilde veri setine gürültü (1) ekliyoruz
y[::10] += 1 * (0.5 - np.random.rand(50))

"""
oluşturduğumuz gürültüleri gözlemlemek için yazdık
"""
plt.figure()
plt.scatter(X,y)
plt.show()
```

```python
#derinliği farklı olan iki karar ağacı oluşturdum
regr_1 = DecisionTreeRegressor(max_depth= 2)
regr_1.fit(X,y)
```

```python
regr_2 = DecisionTreeRegressor(max_depth= 5)
regr_2.fit(X,y)
```

```python
X_test = np.arange(0, 5, 0.05)[:, np.newaxis]
y_pred_1 = regr_1.predict(X_test)
y_pred_2 = regr_2.predict(X_test)
```

```python
#sarı olan bizim eğitmek için kullandığımız veri setimiz biz ona gürültüler ekledik aynı zamanda yeşil ile de plot olarak gösterdik.
plt.figure()
plt.scatter(X, y, c ="yellow" , label ="data")
plt.plot(X, y, c ="green" , label ="data")
plt.xlabel("data")
plt.ylabel("target")
plt.legend()
plt.show()
```

```python
#Derinliği 2 olan modelimi inceledim
plt.figure()
plt.plot(X_test, y_pred_1, c="blue", label="Max Depth: 2")
plt.xlabel("data")
plt.ylabel("target")
plt.legend()
plt.show()
```

```python
#oluşturmuş olduğumuz karar ağaçlarından derinliği 2 olan modelin grafiği aşağıdaki gibidir.
plt.figure()
plt.scatter(X, y, c ="yellow" , label ="data")
plt.plot(X_test, y_pred_1, c="blue", label="Max Depth: 2")
plt.xlabel("data")
plt.ylabel("target")
plt.legend()
plt.show()
```

```python
#Derinliği 5 olan modelimi inceledim
plt.figure()
plt.plot(X_test, y_pred_2, c="red", label="Max Depth: 5")
plt.xlabel("data")
plt.ylabel("target")
plt.legend()
plt.show()
```

```python
#oluşturmuş olduğumuz karar ağaçlarından derinliği 5 olan modelin grafiği aşağıdaki gibidir.
# Grafiktende görüldüğü gibi yaklaşık kararları 5 derinliğe sahip model ile verebiliriz.
plt.figure()
plt.scatter(X, y, c ="yellow" , label ="data")
plt.plot(X_test, y_pred_2, c="red", label="Max Depth: 5")
plt.xlabel("data")
plt.ylabel("target")
plt.legend()
plt.show()
```

```python
#Derinliği arttırdığımda oluşacak değişikliği gözlemlemek için yeni bir tanımlama yapıyorum
regr_3 = DecisionTreeRegressor(max_depth= 20)
regr_3.fit(X,y)
y_pred_3 = regr_3.predict(X_test)
plt.figure()
plt.plot(X_test, y_pred_3, c="red", label="Max Depth: 20")
plt.xlabel("data")
plt.ylabel("target")
plt.legend()
plt.show()
```

```python
#derinliği 20 olan bu modelin veriyi ezberlediği ve veri setindeki gibi bir grafik ortaya çıkattığını söyleyebiliriz.
#Bunu daha nalaşılır görebilmek için SARI grafik verisetindeki veriler olacak şekilde tekrar grafik çiziyorum.

plt.figure()
plt.scatter(X, y, c ="yellow" , label ="data")
plt.plot(X_test, y_pred_3, c="red", label="Max Depth: 20")
plt.xlabel("data")
plt.ylabel("target")
plt.legend()
plt.show()
```

```python
# Bu veri seti üzerine iki tane decision tree inşaa ettik, bu derinlik sınırlamalarında yapabileceğimiz yorum: derinlik arttıkça modelleme 
# daha girintili bir halde olacak ve kullanılan veri setine benzemeye başlayacaktı. Bu da modelin veriyi ezberlediği anlamına gelecektir.
plt.figure()
#plt.scatter(X, y, c ="yellow" , label ="data")
plt.plot(X, y, c ="green" , label ="data")
plt.plot(X_test, y_pred_1, c="blue", label="Max Depth: 2")
plt.plot(X_test, y_pred_2, c="red", label="Max Depth: 5")
plt.xlabel("data")
plt.ylabel("target")
plt.legend()
plt.show()
```
